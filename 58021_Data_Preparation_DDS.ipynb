```python
NewData_df  = DataA.copy()
```

```python
from xlrd.formula import colname
class DataFrame:
  def __init__(self, col, val):
    self.col = colname
    self.val = val

  def make_df(self):
    convert = NewData_df.loc[NewData_df[self.col]==self.val]
    df=pd.DataFrame(convert)
    col_list = df.columns.tolist()
    index = col_list.index(self.col_name)
    new_pos= col_list[index:]+col_list[:index]
    df - df [new_pos]
    return df
```

```python
def rename_cols(old_cols, new_cols):
  for i, old_col in enumerate(old_cols):
    new_col = new_col[i]
    rename_col(old_col,new_col)

def rename_col(old_col, new_col):
  return NewData_df.rename(columns = {old_col:new_col}, inplace = True)
```

```python
def find_list_replace(col_name, find_vals, replace_val):
  for find_val in find_vals:
    find_replace(col_name, find_val, replace_val)

def find_replace(col_name, find_val, replace_val):
  index = NewData_df.index[NewData_df[col_name]==find_val].tolist()
  NewData_df.loc[index,col_name] = replace_val
  return NewData_df.loc[index,col_name]
```

```python
## for deleting entire column
def del_col(col_name):
  return NewData_df.drop(labels=col_name, axis = 1)
```
