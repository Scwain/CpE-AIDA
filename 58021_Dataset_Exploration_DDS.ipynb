```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pov_country = pd.read_csv('wb_subnational_poverty_data.csv')
pov_country
```

```python
pov_country.head()
```

```python
pov_country.info()
```

```python
pov_country.describe()
```

```python
new_pov_country = pov_country.sort_values ('country_code', ascending = False)
new_pov_country
```

```python
imp_data = new_pov_country[['country_code', 'value', 'year']]
imp_data.sort_values('country_code', ascending = True)
```

```python
imp_data.plot(kind = 'line', x='country_code',y='value') 
plt.title ('Value with respect to country code', size=50)
plt.rcParams["axes.labelsize"] = 30
fig = plt.gcf()
fig.set_size_inches(80, 20)
plt.xticks(rotation=90, size=30)
plt.yticks(size=30)
plt.show()
```
